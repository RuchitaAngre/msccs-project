# Data Visualization

# Analyzing how people  find Online Grocery Shopping Easier 

```python
data['If Yes, Do you find Online Grocery Shopping Easier ?'].value_counts().plot(kind='pie',autopct='%1.2f%%',explode=[.05,.05],figsize=(10, 8))
plt.axis('off')
plt.title("If Yes, Do you find Online Grocery Shopping Easier ?")
plt.show()
```

# Analyzing   why people  Shop Online

```python
data[ 'Why do you  Shop Online?  '].value_counts().plot(kind='barh', figsize=(10, 8))
plt.xlabel("Counts of Reasons", labelpad=14)
plt.ylabel("Reasons for Shopping Grocery Online", labelpad=14)
plt.xlim(0, 180)
plt.title("Why do you  Shop Online?", y=1.02);
```

# Analyzing  how often people shop online in last six month

```python
data['How often do you shop online in last six month?'].value_counts().plot(kind='line', figsize=(10, 8))
plt.xlabel("shop online in last six month", labelpad=14)
plt.ylabel("Counts", labelpad=14)
plt.title("How often do you shop online in last six month?", y=1.02);
```

# Analyzing Ratings for factor  On Time Delivery 

```python
sns.violinplot(x=data['On scale of 5 , can you rate for following factor ? [On Time Delivery]'])
plt.title("On Time Delivery", y=1.02);
```

# Analyzing Overall  Online Grocery Shopping Experience of peoples

```python
data['Overall, How would you rate your Online Grocery Shopping Experience'].value_counts().plot(kind='box', figsize=(10, 8))
plt.ylabel("Ratings")
plt.ylim(30, 120)
plt.title("How would you rate your Online Grocery Shopping Experience", y=1.02);
```

# To know how consumer evaluated the factors towards online grocery shopping

```python
plt.hist([data['On scale of 5 , can you rate for following factor ? [On Time Delivery]'],data['On scale of 5 , can you rate for following factor ? [Trustworthy]'],data['On scale of 5 , can you rate for following factor ? [No transaction issues]'],data['On scale of 5 , can you rate for following factor ? [Security]'],data['On scale of 5 , can you rate for following factor ? [Product  Description]'],data['On scale of 5 , can you rate for following factor ? [Product Reviews]']], color=['blue', 'orange','red','green','purple','yellow'])
plt.title('To know how consumer evaluated the factors towards online grocery shopping')
plt.ylabel('Ratings')
plt.xlabel('scale')
plt.legend(['On Time Delivery','Trustworthy','No transaction issues','Security','Product  Description','Product Reviews'],loc=9)
plt.ylim(0, 170)
plt.rcParams["figure.figsize"] = (12, 10)
plt.show()
```

# To identify consumer perception towards online payment

Grouping 2 columns (people find dificulties in online paymnets and how much they rate for factor No transaction issue )

```python
from scipy import stats
def quantile_plot(x, **kwargs):
    quantiles, xr = stats.probplot(x, fit=False)
    plt.scatter(xr, quantiles, **kwargs)

g = sns.FacetGrid(data, col='Do you find any difficulties in online payment ?', height=6)
g.map(quantile_plot, 'On scale of 5 , can you rate for following factor ? [No transaction issues]')
```

# To examine how much consumer shopped online in last six month

Grouping 2 columns (How often people shop online in last six month and their overall experience shopping online)

```python
ordered_days = data['How often do you shop online in last six month?'].value_counts().index
g = sns.FacetGrid(data, row='How often do you shop online in last six month?', row_order=ordered_days,
                  height=2.7, aspect=4)
g.map(sns.kdeplot,'Overall, How would you rate your Online Grocery Shopping Experience' )
```
