# Analyzing would people prefer Shopping Online

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
fig.suptitle('Would People will prefer Shopping Grocery Online ')
g1 = sns.countplot(x='Did you find good quality of product online as compare to your Retailer Shop?',
                  hue='Would you prefer  shopping grocery online  than shopping  from your Retailer Shop?  ', data=data, ax=ax1, order=['Yes', 'No'])

g2 = sns.countplot(x='Do you find enough  Quantity of grocery product  online in comparison to your Retailer ? ',
                  hue='Would you prefer  shopping grocery online  than shopping  from your Retailer Shop?  ', data=data, ax=ax2, order=['Yes', 'No'])

g1.legend_.remove()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., title='Would you prefer')
g1.set_xlabel('Quantity')
g2.set_xlabel('Quality')
g2.set_ylabel('')
plt.show()
```

# 4) To identify consumer perception towards online payment

Grouping 2 columns (people find dificulties in online paymnets and how much they rate for factor No transaction issue )

```python
from scipy import stats
def quantile_plot(x, **kwargs):
    quantiles, xr = stats.probplot(x, fit=False)
    plt.scatter(xr, quantiles, **kwargs)

g = sns.FacetGrid(data, col='Do you find any difficulties in online payment ?', height=6)
g.map(quantile_plot, 'On scale of 5 , can you rate for following factor ? [No transaction issues]')
```
