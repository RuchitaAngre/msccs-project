# Converting 2nd column[If Yes, Do you find Online Grocery Shopping Easier ?] to interger

```python
def trans_con(x):
    if x == 'Yes':
        return 1
    if x == 'No':
        return 0
```

```python
data['If Yes, Do you find Online Grocery Shopping Easier ?'] = data['If Yes, Do you find Online Grocery Shopping Easier ?'].apply(trans_con)
```

```python
data['Did you find good quality of product online as compare to your Retailer Shop?'] = data['Did you find good quality of product online as compare to your Retailer Shop?'].apply(trans_con)
```

# Converting 17th column[How often do you shop online in last six month?] to interger where Everyday=1,Weekly=2,Monthly=3 & 2times a week =4

```python
def trans_con(x):
    if x == 'Everyday':
        return 1
    if x == 'Weekly':
        return 2
    if x == 'Monthly':
        return 3
    if x == '2 times a week':
        return 4
```

```python
data['How often do you shop online in last six month?'] = data['How often do you shop online in last six month?'].apply(trans_con)
```
